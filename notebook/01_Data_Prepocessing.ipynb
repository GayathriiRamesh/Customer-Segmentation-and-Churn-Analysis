## Data Cleaning

```python
# Initial dataset shape
print("Original dataset shape:", df.shape)
```

```python
# Remove rows without CustomerID, since they cannot be used for customer analysis
df = df.dropna(subset=['CustomerID'])
```

```python
# Remove duplicate records to avoid repeated transactions
df = df.drop_duplicates()
```

```python
# Convert InvoiceDate to datetime format for time-based analysis
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Remove rows with negative quantities (usually returned items)
df = df[df["Quantity"] > 0]

# Remove rows with zero or negative prices to ensure valid sales data
df = df[df['UnitPrice'] > 0]
```

```python
# Final dataset shape
print("Cleaned dataset shape:", df.shape)
```

```python
# Save the cleaned dataset to a CSV file
df.to_csv("~/customer_segmentation_churn/data/cleaned_data.csv",index = False)
```
